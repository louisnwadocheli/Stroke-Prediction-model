```python
#splitting data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
xgb=XGBClassifier(objective='binary:logistic', seed=42,n_estimators=200)
xgb.fit(X_train,y_train,verbose=True, early_stopping_rounds=5, eval_metric='auc', eval_set=[(X_test,y_test)])
```

```python
#feature importance using xgboost
```

```python
# Train the model on the training set
xgb.fit(X_train, y_train)

# Make predictions on the testing set
y_pred =xgb.predict(X_test)
```

```python
#valuate the model performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
#from sklearn.metrics import accuracy_score
classification_rep = classification_report(y_test, y_pred)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print accuracy
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
# Get feature importance scores
importance = xgb.get_booster().get_score(importance_type='weight')
print("Feature Importance (Weight):", importance)

importance_gain = xgb.get_booster().get_score(importance_type='gain')
print("Feature Importance (Gain):", importance_gain)

importance_cover = xgb.get_booster().get_score(importance_type='cover')
print("Feature Importance (Cover):", importance_cover)
```
